## Cartesian Circular Task


In this block, we import the library needed for graphs and another one for generating data.

```python
import numpy as np
import matplotlib.pyplot as plt
```

We are defining the parameters of the signal.

```python
T = 3 # Period in second units 
f = 1 / T # Frequency in Hertz units.
w = 2*np.pi*f # Angular frequency in radians per second
```

The time array starts from 0 to 3 seconds, with samples taken every 1 millisecond

```python
time = np.linspace(0,3,3000)
```

We define the parameters for position in x and y

```python
point_x = 0
point_y = 0
```

We build two arrays of coordinates x and y 

```python
x = point_x + np.sin(w*time)
y = point_y + np.cos(w*time)
```

```python
plt.figure(figsize=(5,5))
plt.plot(x,y)
plt.show()
```
